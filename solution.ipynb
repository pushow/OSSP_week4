```python
def rotateString(s, goal):
  len_S = len(s)
  for i in range(len_S):
      a=s[0]
      s=s[1:]
      s+=a
      if s == goal:
        return True
  return False
```

```python
'''
Testcase 1. 
Input: s = "abcde", goal = "cdeab"
Output: True
'''
rotateString("abcde", "cdeab")
```

```python
'''
Testcase 2. 
Input: s = "abcde", goal = "abced"
Output: False
'''
rotateString("abcde", "abced")
```

```python
'''
Testcase 3. 
Input: s = "abcde", goal = "abcde"
Output: True
'''
rotateString("abcde", "abcde")
```
